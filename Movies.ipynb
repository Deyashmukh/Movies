```python
#Missing Data

for col in df.columns:
    pct_missing=(df[col].isnull().mean())
    print("{} - {}%".format(col,pct_missing))
    
```

```python
#changing data types of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
corr_matrix=df.corr()
```

```python
sns.heatmap(corr_matrix,cmap="coolwarm",annot=True)
```

```python
#Assigning random codes to columns whose data type is an object 
df_numerized=df

for col in df_numerized:
    if(df_numerized[col].dtype=="object"):
        df_numerized[col]=df_numerized[col].astype("category")
        df_numerized[col]=df_numerized[col].cat.codes
    
df_numerized
```

```python
corr_matrix=df_numerized.corr()
sns.heatmap(corr_matrix,cmap="coolwarm",annot=True)
```

```python
corr_pairs=corr_matrix.unstack()
corr_pairs
sort_pairs=corr_pairs.sort_values()
sort_pairs=sort_pairs[sort_pairs>0.5]
sort_pairs
```

```python

# Looking at the top 15 companies by gross revenue
grosssum=dfname.groupby("company")[["gross"]].sum()
grosssum
grosssumsort=grosssum.sort_values(by=["gross"],ascending=False).head(15).astype("int64")
grosssumsort

```

```python
grosssum=dfname.groupby(["company","year"])[["gross"]].sum()
grosssum
grosssumsort=grosssum.sort_values(by=["gross"],ascending=False).head(15).astype("int64")
grosssumsort
```
